Using SimpleImputer to fill in the NaN values

Normalisation of each attribute using MinMaxScaler

```python
# Normalising values using min-max scaler for similar range
scaler = MinMaxScaler()
X_normalised = scaler.fit_transform(X_imputed)
```

Changing class values by replacing class1 to 0 and class2 to 1

### Part 1: Cross-validation without parameter tuning

In this section, we apply multiple classifiers to the preprocessed dataset, namely: Logistic Regression, Naïve Bayes, Decision Tree, Bagging, AdaBoost and Gradient Boosting. They are set to perform at random_state = 0 and their performance is evaluated using 10-fold stratified cross validation. Each classifier returns the average cross-validation score from its function.

cvKFold is a 10-fold stratified cross-validation splitter that keeps class ratios consistent across folds, shuffles the data and makes results reproducible.

```python
# Setting the 10 fold stratified cross-validation
cvKFold = StratifiedKFold(n_splits=10, shuffle=True, random_state=0)
# The stratified folds from cvKFold should be provided to the classifiers
```

The below function runs LogisticRegression from sklearn.linear_model and performs cross-validation, returning the average accuracy score across folds.

```python
# Logistic Regression
def logregClassifier(X, y):
    model = LogisticRegression(random_state=0, max_iter=5000)
    scores = cross_val_score(model, X, y, cv=cvKFold)
    return scores.mean()
```

The below function runs GaussianNB from sklearn.naive_bayes and performs cross-validation, returning the average accuracy score across folds.

```python
# Naïve Bayes
def nbClassifier(X, y):
    model = GaussianNB()
    scores = cross_val_score(model, X, y, cv=cvKFold)
    return scores.mean()
```

The below function runs DecisionTreeClassifier from sklearn.tree with entropy criterion and performs cross-validation, returning the average accuracy score across folds.

```python
# Decision Tree
def dtClassifier(X, y):
    model = DecisionTreeClassifier(criterion="entropy", random_state=0)
    scores = cross_val_score(model, X, y, cv=cvKFold)
    return scores.mean()
```

The below function runs BaggingClassifier from sklearn.ensemble using the base tree DecisionTreeClassifier with entropy criterion and performs cross-validation, returning the average accuracy score across folds.

```python
# Ensembles: Bagging
def bagDTClassifier(X, y, n_estimators, max_samples, max_depth):
    base_tree = DecisionTreeClassifier(criterion="entropy", max_depth=max_depth, random_state=0)
    model = BaggingClassifier(
        estimator=base_tree,
        n_estimators=n_estimators,
        max_samples=max_samples,
        random_state=0
    )
    scores = cross_val_score(model, X, y, cv=cvKFold)
    return scores.mean()
```

The below function runs AdaBoostClassifier from sklearn.ensemble using the base tree DecisionTreeClassifier with entropy criterion and performs cross-validation, returning the average accuracy score across folds.

```python
# Ensembles: AdaBoost
def adaDTClassifier(X, y, n_estimators, learning_rate, max_depth):
    base_tree = DecisionTreeClassifier(criterion="entropy", max_depth=max_depth, random_state=0)
    model = AdaBoostClassifier(
        estimator=base_tree,
        n_estimators=n_estimators,
        learning_rate=learning_rate,
        random_state=0
    )
    scores = cross_val_score(model, X, y, cv=cvKFold)
    return scores.mean()
```

The below function runs GradientBoostingClassifier from sklearn.ensemble and performs cross-validation, returning the average accuracy score across folds.

```python
# Ensembles: Gradient Boosting
def gbClassifier(X, y, n_estimators, learning_rate):
    model = GradientBoostingClassifier(
        n_estimators=n_estimators,
        learning_rate=learning_rate,
        random_state=0
    )
    scores = cross_val_score(model, X, y, cv=cvKFold)
    return scores.mean()
```

### Part 1 Results

The results of executing each classifier is listed below. Bagging Classifier produces the highest average cross-validation score, while Decision Tree Classifier produces the lower average score on the Rice Classification dataset.

```python
# Parameters for Part 1:

#Bagging
bag_n_estimators = 50
bag_max_samples = 100
bag_max_depth = 5

#AdaBoost
ada_n_estimators = 50
ada_learning_rate = 0.5
ada_bag_max_depth = 5

#Gradient Boosting
gb_n_estimators = 50
gb_learning_rate = 0.5

# Print results for each classifier in part 1 to 4 decimal places here:
logR_accuracy = logregClassifier(X, y)
print("LogR average cross-validation accuracy: {:.4f}".format(logR_accuracy))
nb_accuracy = nbClassifier(X, y)
print("NB average cross-validation accuracy: {:.4f}".format(nb_accuracy))
dt_accuracy = dtClassifier(X, y)
print("DT average cross-validation accuracy: {:.4f}".format(dt_accuracy))
bagging_accuracy = bagDTClassifier(X, y, bag_n_estimators, bag_max_samples, bag_max_depth)
print("Bagging average cross-validation accuracy: {:.4f}".format(bagging_accuracy))
adaBoost_accuracy = adaDTClassifier(X, y, ada_n_estimators, ada_learning_rate, ada_bag_max_depth)
print("AdaBoost average cross-validation accuracy: {:.4f}".format(adaBoost_accuracy))
gb_accuracy = gbClassifier(X, y, gb_n_estimators, gb_learning_rate)
print("GB average cross-validation accuracy: {:.4f}".format(gb_accuracy))
```

### Box Plot of Cross Validation Scores

```python
# Aggregating the models
algorithms = {
    'Logistic Regression': LogisticRegression(random_state=0, max_iter=5000),
    'Naive Bayes': GaussianNB(),
    'Decision Tree': DecisionTreeClassifier(criterion="entropy", random_state=0),
    'Bagging': BaggingClassifier(DecisionTreeClassifier(criterion="entropy", max_depth=5, random_state=0), 
                                n_estimators=50, max_samples=100, random_state=0),
    'AdaBoost': AdaBoostClassifier(DecisionTreeClassifier(criterion="entropy", max_depth=5, random_state=0),
                                  n_estimators=50, learning_rate=0.5, random_state=0),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=50, learning_rate=0.5, random_state=0)
}

# Get CV scores for each algorithm
cv_results = {}
for name, model in algorithms.items():
    scores = cross_val_score(model, X, y, cv=cvKFold, scoring='accuracy')
    cv_results[name] = scores

# Box plot of the 5 models
plt.figure(figsize=(10,6))
plt.boxplot(cv_results.values(), tick_labels=cv_results.keys())
plt.ylabel('Cross-Validation Accuracies')
plt.title('CV Score Distribution Across Algorithms')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Part 2: Cross-validation with parameter tuning

In this section, we apply two other classifiers to the preprocessed dataset, namely: K Nearest Neighbor and Random Forest. The dataset is first split into training and test subsets using train_test_split from sklearn.model_selection. They are set to perform at random_state = 0 and their performance is evaluated using 10-fold stratified cross validation. Each classifier returns the average cross-validation score from its function.

Splitting the dataset into 75% train data and 25% test data

```python
# Splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=0)
```

The below function runs KNeighborsClassifier from sklearn.neighbors using GridSearchCV by tuning the hyperparameters n_neighbors and p to find the best combination, returning best hyperparamters, cross-validation accuracy and test accuracy.

```python
# KNN
k = [1, 3, 5, 7]
p = [1, 2]

def bestKNNClassifier(X, y):
 
    param_grid = {
        "n_neighbors": k,
        "p": p
    }
    
    grid = GridSearchCV(
        KNeighborsClassifier(),
        param_grid,
        cv=cvKFold,
        scoring="accuracy"
    )
    
    grid.fit(X_train, y_train)
    
    best_parameters = grid.best_params_
    best_cv_accuracy = grid.best_score_
    
    y_pred = grid.best_estimator_.predict(X_test)
    test_accuracy = accuracy_score(y_test, y_pred)
    
    return best_parameters, best_cv_accuracy, test_accuracy
```

The below function runs RandomForestClassifier from sklearn.ensemble using GridSearchCV by tuning the hyperparameters n_estimators and max_leaf_nodes to find the best combination, returning best hyperparamters, cross-validation accuracy, test accuracy, macro average F1 score and the weighted average F1 score.

```python
# Random Forest
# You should use RandomForestClassifier from sklearn.ensemble with information gain and max_features set to ‘sqrt’.
n_estimators = [10, 30, 60, 100]
max_leaf_nodes = [6, 12]

def bestRFClassifier(X, y):
        
    param_grid = {
        "n_estimators": n_estimators,
        "max_leaf_nodes": max_leaf_nodes
    }
    
    rf = RandomForestClassifier(
        criterion="entropy",
        max_features="sqrt",
        random_state=0
    )
    
    grid = GridSearchCV(
        rf,
        param_grid,
        cv=cvKFold,
        scoring="accuracy"
    )
    
    grid.fit(X_train, y_train)
    
    best_parameters = grid.best_params_
    best_cv_accuracy = grid.best_score_
    
    y_pred = grid.best_estimator_.predict(X_test)
    test_accuracy = accuracy_score(y_test, y_pred)
    
    f1_macro = f1_score(y_test, y_pred, average="macro")
    f1_weighted = f1_score(y_test, y_pred, average="weighted")
    
    return best_parameters, best_cv_accuracy, test_accuracy, f1_macro, f1_weighted
```

### Part 2: Results

The results of executing each classifier is listed below. Random Forest Classifier outperforms KNN slightly better in the cross-validation and test accuracies on the Rice Classification dataset. F1 scores for Random Forest indicate balanced performance across classes.

```python
# Perform Grid Search with 10-fold stratified cross-validation (GridSearchCV in sklearn). 
# The stratified folds from cvKFold should be provided to GridSearchCV

# This should include using train_test_split from sklearn.model_selection with stratification and random_state=0
# Print results for each classifier here. All results should be printed to 4 decimal places except for
# "k", "p", n_estimators" and "max_leaf_nodes" which should be printed as integers.
knn_parameters, knn_cv_accuracy, knn_test_accuracy = bestKNNClassifier(X, y)
print("KNN best k:", knn_parameters["n_neighbors"])
print("KNN best p:", knn_parameters["p"])
print("KNN cross-validation accuracy: {:.4f}".format(knn_cv_accuracy))
print("KNN test set accuracy: {:.4f}".format(knn_test_accuracy))

print()

rf_parameters, rf_cv_accuracy, rf_test_accuracy, rf_f1_macro, rf_f1_weighted = bestRFClassifier(X, y)
print("RF best n_estimators:", rf_parameters["n_estimators"])
print("RF best max_leaf_nodes:", rf_parameters["max_leaf_nodes"])
print("RF cross-validation accuracy: {:.4f}".format(rf_cv_accuracy))
print("RF test set accuracy: {:.4f}".format(rf_test_accuracy))
print("RF test set macro average F1: {:.4f}".format(rf_f1_macro))
print("RF test set weighted average F1: {:.4f}".format(rf_f1_weighted))
```

### Confusion Matrix Random Forest

```python
from sklearn.metrics import confusion_matrix

# Create model with best parameters (n_est=30, max_leaf=12 from your results)
rf_model = RandomForestClassifier(criterion="entropy", max_features="sqrt", 
                                  n_estimators=30, max_leaf_nodes=12, random_state=0)
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)

# Plot confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Random Forest')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# Parameter sensitivity 
# k = [1, 3, 5, 7] and p = [1, 2] already defined
# n_estimators = [10, 30, 60, 100] and max_leaf_nodes = [6, 12] already defined
# cvkFold already defined

# Parameter sensitivity for KNN
knn_results = {}
for pvalues in p:
    scores = []
    for kvalues in k:
        knn = KNeighborsClassifier(n_neighbors=kvalues, p=pvalues)
        cv_scores = cross_val_score(knn, X_train, y_train, cv=cvKFold, scoring='accuracy')
        scores.append(cv_scores.mean())
    knn_results[f'p={pvalues}'] = scores

# Parameter sensitivity for Random Forest
rf_results = {}
for max_leaf in max_leaf_nodes:
    scores = []
    for n_est in n_estimators:
        rf = RandomForestClassifier(
            criterion="entropy", 
            max_features="sqrt", 
            n_estimators=n_est, 
            max_leaf_nodes=max_leaf, 
            random_state=0
        )
        cv_scores = cross_val_score(rf, X_train, y_train, cv=cvKFold, scoring='accuracy')
        scores.append(cv_scores.mean())
    rf_results[f'max_leaf={max_leaf}'] = scores

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# KNN Parameter Sensitivity Plot
for label, scores in knn_results.items():
    ax1.plot(k, scores, marker='o', label=label, linewidth=2, markersize=6)
ax1.set_xlabel('k (Number of Neighbors)')
ax1.set_ylabel('Cross-Validation Accuracy')
ax1.set_title('KNN Parameter Sensitivity')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_xticks(k)

# Random Forest Parameter Sensitivity Plot
for label, scores in rf_results.items():
    ax2.plot(n_estimators, scores, marker='s', label=label, linewidth=2, markersize=6)
ax2.set_xlabel('Number of Estimators')
ax2.set_ylabel('Cross-Validation Accuracy')
ax2.set_title('Random Forest Parameter Sensitivity')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('param_sensitivity.png', dpi=300, bbox_inches='tight')
plt.show()
```
